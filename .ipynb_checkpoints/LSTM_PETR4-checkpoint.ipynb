```python
# === 1. IMPORTAÇÕES ===
```

```python
import pandas as pd
import numpy as np
import cloudpickle
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
```

```python
# === 2. PRÉ-PROCESSAMENTO E ENGENHARIA DE FEATURES ===
```

```python
def preprocess_dataframe(df):
    df["Data"] = pd.to_datetime(df["Data"])
    for col in ["Fechamento", "Abertura", "Maxima", "Minima"]:
        df[col] = df[col].astype(str).str.replace(",", ".").astype(float)
    def parse_volume(val):
        val = str(val).replace(".", "").replace(",", ".")
        return float(val.replace("K", "")) * 1e3 if "K" in val else float(val.replace("M", "")) * 1e6
    df["Volume"] = df["Volume"].apply(parse_volume)
    df["Variacao"] = df["Variacao"].astype(str).str.replace(",", ".").str.replace("%", "").astype(float)
    return df.sort_values("Data").reset_index(drop=True)

def add_features_lstm(df):
    df["mm5"] = df["Fechamento"].rolling(window=5).mean()
    df["mm10"] = df["Fechamento"].rolling(window=10).mean()
    df["retorno_dia"] = df["Fechamento"].pct_change() * 100
    df["volatilidade"] = df["Fechamento"].rolling(window=5).std()
    return df.dropna().reset_index(drop=True)

def scale_and_window(df, window_size=10, target_col="Fechamento"):
    features = ["Abertura", "Maxima", "Minima", "Volume", "Variacao", "mm5", "mm10", "retorno_dia", "volatilidade"]
    scaler = StandardScaler()
    scaled = scaler.fit_transform(df[features + [target_col]])
    X, y = [], []
    for i in range(window_size, len(df)):
        X.append(scaled[i-window_size:i, :-1])
        y.append(scaled[i, -1])
    return np.array(X), np.array(y), scaler
```

```python
# === 3. CONSTRUÇÃO DO MODELO ===
```

```python
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.3))
    model.add(LSTM(64, return_sequences=True))  
    model.add(Dropout(0.3))
    model.add(LSTM(64, return_sequences=True))  
    model.add(Dropout(0.3))
    model.add(LSTM(32))                         
    model.add(Dropout(0.2))
    model.add(Dense(1))                         # Previsão do fechamento

    model.compile(optimizer=Adam(0.0001), loss='mean_squared_error')
    return model
```

```python
# === 4. LEITURA E PROCESSAMENTO DOS DADOS ===
```

```python
train_raw = pd.read_csv("PETR4_train.csv")
test_raw = pd.read_csv("PETR4_test.csv")

train = add_features_lstm(preprocess_dataframe(train_raw))
test = add_features_lstm(preprocess_dataframe(test_raw))

X_train, y_train, scaler = scale_and_window(train)
X_test, y_test, _ = scale_and_window(test)
```

```python
# === 5. CALLBACK DE EARLY STOPPING ===
```

```python
early_stop = EarlyStopping(
    monitor='loss',             # Pode ser 'val_loss' se quiser generalização
    patience=30,
    verbose=1,
    mode='min',
    restore_best_weights=True   # Recupera os melhores pesos
)
```

```python
# === 6. TREINAMENTO DO MODELO COM EARLY STOPPING ===
```

```python
model = build_lstm_model(X_train.shape[1:])
history = model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stop],
    verbose=1
)
```

```python
# === 7. PREDIÇÃO + INVERSÃO DA ESCALA ===
```

```python
y_pred_scaled = model.predict(X_test)
last_column_index = -1
```

```python
# Criar matriz dummy para fazer o inverse_transform
y_test_inv = scaler.inverse_transform(np.concatenate([X_test[:, -1], y_test.reshape(-1, 1)], axis=1))[:, last_column_index]
y_pred_inv = scaler.inverse_transform(np.concatenate([X_test[:, -1], y_pred_scaled], axis=1))[:, last_column_index]
```

```python
# === 8. VISUALIZAÇÃO ===
plt.figure(figsize=(14, 6))
plt.plot(test["Data"][-len(y_test):], y_test_inv, label="Real", linewidth=2)
plt.plot(test["Data"][-len(y_pred_scaled):], y_pred_inv, label="Previsto", linestyle="--")
plt.title("📊 Previsão do Preço de Fechamento PETR4 com LSTM")
plt.xlabel("Data")
plt.ylabel("Preço (R$)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ==Bloco de Métricas Quantitativa==
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Cálculo das métricas com os dados invertidos
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
mae = mean_absolute_error(y_test_inv, y_pred_inv)
r2 = r2_score(y_test_inv, y_pred_inv)
```

```python
print("📊 Avaliação Quantitativa do Modelo LSTM")
print(f"RMSE (Root Mean Squared Error): {rmse:.4f}")
print(f"MAE  (Mean Absolute Error):      {mae:.4f}")
print(f"R²   (Coeficiente de Determinação): {r2:.4f}")
```

```python
#== Exportar Resultados como CSV ==
```

```python
# Criar DataFrame com resultados
resultados_df = pd.DataFrame({
    "Data": test["Data"].values[-len(y_pred_inv):],
    "Preco_Real": y_test_inv,
    "Preco_Previsto": y_pred_inv,
    "Erro_Absoluto": np.abs(y_test_inv - y_pred_inv),
    "Erro_Percentual": np.abs((y_test_inv - y_pred_inv) / y_test_inv) * 100
})

# Exportar para CSV
resultados_df.to_csv("resultados_petr4_lstm.csv", index=False)

print("✅ Arquivo 'resultados_petr4_lstm.csv' salvo com sucesso.")
```

```python
# === 7. SALVANDO MODELO E SCALER PARA USO NO APP ===
```

```python
# Salvar modelo treinado
with open("modelo_lstm.pkl", "wb") as f:
    cloudpickle.dump(model, f)

# Salvar scaler para uso na normalização da previsão futura
with open("scaler_lstm.pkl", "wb") as f:
    cloudpickle.dump(scaler, f)

print("✅ Modelo e scaler salvos com sucesso!")
```
